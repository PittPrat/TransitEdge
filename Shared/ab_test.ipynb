```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
baseline = np.random.normal(45, 5, 100)
optimized = baseline * np.random.uniform(0.7, 0.9, 100)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(baseline, alpha=0.6, label='Baseline', bins=15)
plt.hist(optimized, alpha=0.6, label='Optimized', bins=15)
plt.xlabel("Travel Time (min)")
plt.ylabel("Frequency")
plt.title("A/B Test: Baseline vs Optimized Travel Time")
plt.legend()
plt.grid(True)
```

```python
plt.savefig("ab_histogram.png")
plt.show()
```
